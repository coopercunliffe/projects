```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ufo.csv')
```

```python
us = df[df['country'] == 'us']
```

```python
us_sitings = us['state'].value_counts()
type(us_sitings)
```

```python
sns.set(rc={'figure.figsize':(18,8)})
sns.barplot(x=us_sitings.index, y=us_sitings)
```

```python
us_sitings = pd.DataFrame(us_sitings)
```

```python
pop = pd.read_csv('pop.csv')
```

```python
us_sitings['pop'] = list(pop['pop'])
```

```python
us_sitings = us_sitings.rename(columns = {'state':'sitings'})
```

```python
us_sitings['per_person'] = us_sitings['sitings'] / us_sitings['pop'] 
```

```python
us_sitings['per_ten_thousand'] = us_sitings['per_person'] * 10000
```

```python
us_sitings = us_sitings.sort_values('per_ten_thousand', ascending=False)
```

```python
sns.barplot(x=us_sitings.index, y=us_sitings['per_ten_thousand'])
```
